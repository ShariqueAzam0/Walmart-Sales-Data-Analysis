```python
# 1st step: Data Exploration & Loading

# We will load the datsets
df = pd.read_csv('Raw_Data.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("Raw Data Shape:", df.shape)
```

```python
df.dtypes
```

```python
# 2nd step: Data cleaning 

df['Category'] = df['Category'].str.strip()
```

```python
finance_columns = ['Unit_Price', 'Discount_Amount', 'Shipping_Cost']
for col in finance_columns:
    df[col] = df[col].astype(str).str.replace('$', '', regex=False).astype(float)
# by this we removed '$' and coverted the price which was object data type to float data type
```

```python
df.dtypes
```

```python
# 3rd step: Data Transformation (We will be Adding new columns logically)

df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d-%m-%Y', errors='coerce')
df['Order_Year'] = df['Order_Date'].dt.year
df['Order_Quarter'] = 'Q' + df['Order_Date'].dt.quarter.astype(str)
df['Order_Month'] = df['Order_Date'].dt.month_name().str[:3] 
df['Day_Of_Week'] = df['Order_Date'].dt.day_name()
```

```python
df.head()
```

```python
df['Base_Price'] = df['Quantity'] * df['Unit_Price']
```

```python
df['Net_Sales'] = df['Base_Price'] - df['Discount_Amount']
```

```python
df['Tax'] = df['Net_Sales'] * 0.08 
# I Take 8% standardized tax rate across all item to keep data model simple
```

```python
df['Total_Amount'] = df['Net_Sales'] + df['Tax'] + df['Shipping_Cost']
```

```python
df[['Base_Price', 'Net_Sales', 'Tax', 'Total_Amount']] = df[['Base_Price', 'Net_Sales', 'Tax', 'Total_Amount']].round(2)
# Rounding these columns to 2 decimal places for neatness
```

```python
# 4th step: Data Restructuring

df['Order_Status'] = df['Order_Status'].replace({'Processing': 'Delivered'})
```

```python
df.columns
```

```python
new_column_order = ['Order_Id', 'Order_Date', 'Order_Year', 'Order_Quarter', 'Order_Month', 'Day_Of_Week', 
                    'Customer_Id', 'Customer_Type', 'Gender', 'City', 'State', 'Country',
                    'Category', 'Product_Name', 'Quantity', 'Unit_Price', 'Base_Price',
                    'Discount_Amount', 'Tax', 'Shipping_Cost', 'Net_Sales', 'Total_Amount', 'Profit_Margin',
                    'Payment_Method', 'Delivery_Method', 'Order_Status'] 

df = df[new_column_order]
```

```python
df.head()
```

```python
# 5th step: saving the file and formating it for mysql compatiblility

df.columns = df.columns.str.lower()
```

```python
df.to_csv('Walmart_Cleaned_Data.csv', index=False)
print("Data cleaning complete. Saved as Walmart_Cleaned_Data.csv")
```
